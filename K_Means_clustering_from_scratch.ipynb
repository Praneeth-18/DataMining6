The K Means algorithm is a type of unsupervised learning, meaning it doesn't require pre-labeled training data. Instead, it operates directly on the actual dataset. Its primary function is to group similar data points into clusters without making predictions.

Here's how K Means Clustering functions:

Iterative Nature: The K Means algorithm iteratively refines the positions of cluster centroids. It starts by choosing 'k' centroids randomly (where 'k' is a predetermined number of clusters) and then iteratively updates these centroids based on the data.

Dataset and Clusters: The algorithm takes a dataset with 'n' data points and a parameter 'k', which is the desired number of clusters. It outputs 'k' centroids, each representing a cluster, and assigns each data point in the dataset to one of these clusters.

Distance Calculation and Assignment: It calculates the distance between each data point and each centroid. Data points are assigned to the cluster represented by the nearest centroid.

Centroid Recalculation: The centroids are recalculated after the initial assignment, typically as the arithmetic mean of all data points in the cluster.

Convergence: This process of assigning points to clusters and recalculating centroids is repeated until the changes in centroid positions are minimal, indicating that the clusters are relatively stable and the algorithm has converged.

In summary, K Means is a straightforward yet powerful algorithm for clustering data based on similarities, without requiring any prior knowledge or training data.

```python
from sklearn.cluster import KMeans
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df.head()
```

```python
plt.scatter(df.Feature_2,df['Feature_1'])
plt.xlabel('Feature_2')
plt.ylabel('Feature_1')
```

## **Encode Categorical Values**

```python
km = KMeans(n_clusters=3, n_init=10)
y_predicted = km.fit_predict(df[['Feature_2','Feature_1']])
y_predicted
```

```python
df['cluster']=y_predicted
df.head()
```

```python
km.cluster_centers_
```

```python
plt.plot(km.cluster_centers_)
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
plt.scatter(df1.Feature_2,df1['Feature_1'],color='green')
plt.scatter(df2.Feature_2,df2['Feature_1'],color='red')
plt.scatter(df3.Feature_2,df3['Feature_1'],color='black')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.xlabel('Feature 2')
plt.ylabel('Feature 1')
plt.legend()
```

```python
scaler = MinMaxScaler()

scaler.fit(df[['Feature_1']])
df['Feature_1'] = scaler.transform(df[['Feature_1']])

scaler.fit(df[['Feature_2']])
df['Feature_2'] = scaler.transform(df[['Feature_2']])
```

```python
df.head()
```

```python
km = KMeans(n_clusters=3, n_init=10)
y_predicted = km.fit_predict(df[['Feature_2','Feature_1']])
y_predicted
```

```python
df['cluster']=y_predicted
df.head()
```

```python
km.cluster_centers_
```

```python
df1 = df[df.cluster==0]
df2 = df[df.cluster==1]
df3 = df[df.cluster==2]
plt.scatter(df1.Feature_2,df1['Feature_1'],color='green')
plt.scatter(df2.Feature_2,df2['Feature_1'],color='red')
plt.scatter(df3.Feature_2,df3['Feature_1'],color='black')
plt.scatter(km.cluster_centers_[:,0],km.cluster_centers_[:,1],color='purple',marker='*',label='centroid')
plt.xlabel('Feature 2')
plt.ylabel('Feature 1')
plt.legend()
```

## **Elbow Plot**

```python
sse = []
k_rng = range(1,10)
for k in k_rng:
    km = KMeans(n_clusters=k, n_init=10)
    km.fit(df[['Feature_2','Feature_1']])
    sse.append(km.inertia_)
```

```python
plt.xlabel('K')
plt.ylabel('Sum of squared error')
plt.plot(k_rng,sse)
```
